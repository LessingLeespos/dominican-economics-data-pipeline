```python
from bs4 import BeautifulSoup
import requests, os
import time
import sqlite3
```

```python
#Database Setup
db_path = "../sql/bancentral_homepage_files.db"
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

#Create table to keep track of URLs and local paths
cursor.execute('''
    CREATE TABLE IF NOT EXISTS downloads(
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        url TEXT UNIQUE,
        filename TEXT,
        download_time TIMESTAMP DEFAULT CURRENT_TIMESTAMP
        ) ''')
conn.commit()
```

```python
url = "https://www.bancentral.gov.do/"
count = 0

file_type = (".pdf","xlsx",".xls",".xlsm",".csv")
html = requests.get(url).text
soup = BeautifulSoup(html, "html.parser")

for link in soup.find_all("a", href=True):
    count += 1
    if link["href"].endswith(file_type):
        n_url = link["href"]
        filename = os.path.basename(n_url.split("?")[0])  # clean file url
        # Check Database
        cursor.execute("SELECT url FROM downloads WHERE url = ?", (n_url,))
        prev_downloaded = cursor.fetchone()
        if prev_downloaded: continue
        #Download New Files
        try:
            dw_data = requests.get(n_url).content
           # RAW Files storage location
            with open(f"C:/Users/Admin/Desktop/Porttfolio/Excel and Power BI/Dominican Republic Economic Data JUN 2026/raw/{filename}", "wb") as f:
               f.write(dw_data)
            print(f"\rDownloaded {filename} from link #{count}", end="")
            time.sleep(0.1)
        except Exception as x:
            print(f"\nError downloading link #{count}: {x}")
print("\nScraping session finished.")
conn.close()
```
