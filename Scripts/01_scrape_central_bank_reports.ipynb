```python
from bs4 import BeautifulSoup
import requests, os
import time
```

```python
url = "https://www.bancentral.gov.do/"
count = 0  # Count iterations

html = requests.get(url).text
soup = BeautifulSoup(html, "html.parser")
for link in soup.find_all("a", href=True):
    count += 1
    if link["href"].endswith(".pdf"):
        n_url = link["href"]
        filename = os.path.basename(n_url.split("?")[0])  # clean file url
        dw_data = requests.get(n_url).content
        
       # RAW Files storage location
        with open(f"C:/Users/Admin/Desktop/Porttfolio/Excel and Power BI/Dominican Republic Economic Data JUN 2026/raw/a00{count}_{filename}", "wb") as f:
           f.write(dw_data)
        print(f"\rDownloaded {filename} from link #{count}", end="")
        time.sleep(1)

```
